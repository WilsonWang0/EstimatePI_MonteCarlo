```python
import random
import matplotlib.pyplot as plt

def monte_carlo_estimate_pi(total_points):

    # Initialize lists for points
    inside_x = []
    inside_y = []
    outside_x = []
    outside_y = []
    point_in = 0
    
    # Generate random points
    for _ in range(total_points):
        rand_x = random.uniform(-1, 1)
        rand_y = random.uniform(-1, 1)

        # Distance from origin
        if (rand_x**2 + rand_y**2) <= 1:  # Inside the circle
            point_in += 1
            inside_x.append(rand_x)
            inside_y.append(rand_y)
            
        else:  # Outside the circle
            outside_x.append(rand_x)
            outside_y.append(rand_y)

    #estimated value of pi
    print(4*point_in / total_points)
    
   
    # Plot the square and the circle
    plt.figure(figsize=(6, 6))
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.gca().set_aspect("equal", adjustable="box")

    # Draw the circle
    circle = plt.Circle((0, 0), 1, color='blue', fill=False)
    plt.gca().add_artist(circle)

    # Plot the points
    plt.scatter(inside_x, inside_y, color="green", s=1, label="Inside Circle")
    plt.scatter(outside_x, outside_y, color="red", s=1, label="Outside Circle")

    # Add labels and legend
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title(f"Monte Carlo Simulation with {total_points} Points")
    plt.legend()
    plt.show()
```

```python
# Example usage
monte_carlo_estimate_pi(1000)
```
